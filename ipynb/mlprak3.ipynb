```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

# Load dataset
df = pd.read_csv('bad_driver.csv')
```

```python
xpoints = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted']
ypoints = df['Number of drivers involved in fatal collisions per billion miles']
plt.scatter(xpoints, ypoints, marker = 'o') 
plt.title("Plot Relasi") 
plt.ylabel("Jumlah Driver") 
plt.xlabel("Persentase Driver") 
plt.show() 
```

```python
plt.figure(figsize=(14, 10))  

# plot 1 
x1 = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding']
y1 = df['Losses incurred by insurance companies for collisions per insured driver ($)']
plt.subplot(2, 4, 1)  
plt.scatter(x1, y1, color='purple')
plt.xlabel('Speeding Fatal Collisions (%)')
plt.ylabel('Insurance Losses ($)')
plt.title('Plot 1')
  
# plot 2 
xx = df['Number of drivers involved in fatal collisions per billion miles']
y2 = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired']
y3 = df['Losses incurred by insurance companies for collisions per insured driver ($)']
plt.subplot(2, 4, 2)  
plt.scatter(xx, y2, color='blue', label='Alcohol-Impaired (%)')
plt.scatter(xx, y3, color='red', label='Insurance Losses ($)')
plt.xlabel('Fatal Collisions per Billion Miles')
plt.ylabel('Value')
plt.title('Scatter 2')
plt.legend()

# plot 3
xhistogram = df['Number of drivers involved in fatal collisions per billion miles']
plt.subplot(2, 4, 3)  
plt.hist(xhistogram, color='green', edgecolor='black')
plt.xlabel('Fatal Collisions per Billion Miles')
plt.ylabel('Frequency')
plt.title('Histogram 3')

# plot 4
xboxplot4 = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Speeding']
plt.subplot(2, 4, 4)  
plt.boxplot(xboxplot4)
plt.title('Box Plot 4')

# plot 5
xboxplot5 = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Alcohol-Impaired']
plt.subplot(2, 4, 5)  
plt.boxplot(xboxplot5)
plt.title('Box Plot 5')

# plot 6
xboxplot6 = df['Percentage Of Drivers Involved In Fatal Collisions Who Were Not Distracted']
plt.subplot(2, 4, 6)  
plt.boxplot(xboxplot6)
plt.title('Box Plot 6')

# plot 7
xhistogram7 = df['Percentage Of Drivers Involved In Fatal Collisions Who Had Not Been Involved In Any Previous Accidents']
plt.subplot(2, 4, 7)  
plt.hist(xhistogram7, color='orange', edgecolor='black')
plt.xlabel('No Previous Accidents (%)')
plt.ylabel('Frequency')
plt.title('Histogram 7')

# plot 8
xhistogram8 = df['Car Insurance Premiums ($)']
plt.subplot(2, 4, 8)  
plt.hist(xhistogram8, color='cyan', edgecolor='black')
plt.xlabel('Car Insurance Premiums ($)')
plt.ylabel('Frequency')
plt.title('Histogram 8')

plt.tight_layout()
plt.show() 
```

```python
pengamatan = df['Losses incurred by insurance companies for collisions per insured driver ($)']
df = pd.DataFrame(pengamatan) 

q = df.quantile(q=[0.25, 0.50, 0.75]).to_numpy() 
q1, q2, q3 = q[0], q[1], q[2] 

print("Nilai Q1: ", q1) 
print("Nilai Q2: ", q2) 
print("Nilai Q3: ", q3) 

iqr = q3 - q1 
print("Nilai IQR: ", iqr) 

whisker = 1.5 * iqr 
print("Nilai Whisker: ", whisker) 

outlier_bawah = q1 - whisker 
outlier_atas = q3 + whisker 

extrim_bawah = q1 - (3 * iqr) 
extrim_atas = q3 + (3 * iqr) 

print("Elemen pada box mulai dari: ", q1, " sampai: ", q3) 
print("Whisker bagian bawah mulai dari: ", outlier_bawah, " sampai: ", q1) 
print("Whisker bagian atas mulai dari: ", q3, " sampai: ", outlier_atas) 

print("Outlier bagian bawah mulai: ", extrim_bawah, " sampai: ", outlier_bawah) 
print("Outlier bagian atas mulai: ", outlier_atas, " sampai: ", extrim_atas) 

dt_outlier_bawah = [] 
for data in pengamatan: 
    if data < outlier_bawah and data > extrim_bawah: 
        dt_outlier_bawah.append(data) 

print("Elemen outlier bawah: ", dt_outlier_bawah) 

dt_outlier_atas = [] 
for data in pengamatan: 
    if data > outlier_atas and data < extrim_atas: 
        dt_outlier_atas.append(data) 

print("Elemen outlier atas: ", dt_outlier_atas) 

dt_extrim_bawah = [] 
for data in pengamatan: 
    if data < extrim_bawah: 
        dt_extrim_bawah.append(data) 

print("Elemen extrim bawah: ", dt_extrim_bawah) 

dt_extrim_atas = [] 
for data in pengamatan: 
    if data > extrim_atas: 
        dt_extrim_atas.append(data) 

print("Elemen extrim atas: ", dt_extrim_atas) 

fig, ax = plt.subplots() 
ax.boxplot(pengamatan) 
plt.show()
```
